- `DataStore` class: Organizes and stores NBA data for predictions.

- **Data Structures**:
  - `gameResults` (List): Stores each game result.
  - `teamStatistics` (List): Stores team stats in string form.
  - `teamDataMap` (HashMap): Maps each team name to its yearly records.
  - `gameResultQueue` (Queue): Holds recent game results for quick access.
  - `teamRanking` (PriorityQueue): Ranks teams by seeding score, for predictions.
  - `headToHead` (2D Array): Tracks results of head-to-head matchups.

- **Methods**:
  - `addGameResult`: Adds a new game result to `gameResults` and `gameResultQueue`.
  - `addTeamStatistic`: Adds a team statistic to `teamStatistics`.
  - `addTeamRecord`: Adds a yearly record for a team in `teamDataMap`.
  - `updateHeadToHead`: Updates head-to-head matchup results.
  - `displayTeamData`: Shows stored data for a specific team.
  - `displayRankedTeams`: Shows all teams ranked by seeding score.

- **Inner Classes**:
  - `TeamRecord`: Holds each team's rank, stat, and year.
  - `TeamData`: Holds a team’s name, seeding score, points, wins, and losses.

- **Main Function**:
  - Loads team data from CSV files in `Cleaned_Data` folder.
  - Reads files by category, processes each file, and stores team records.
  - Displays data for a sample team ("Phoenix").


```python
%%writefile DataStore.java
import java.util.*;
import weka.core.Instances;
import java.io.File;
import weka.core.Instance;
import weka.core.converters.ConverterUtils.DataSource;

public class DataStore {

    // List to store game results and team statistics
    private List<String> gameResults;
    private List<String> teamStatistics;

    // HashMap for quick lookups of team-specific data, using List<TeamRecord> to store data per year
    private HashMap<String, List<TeamRecord>> teamDataMap;

    // Queue to handle real-time game result inputs
    private Queue<String> gameResultQueue;

    // PriorityQueue to rank teams based on predicted seeding scores
    private PriorityQueue<TeamData> teamRanking;

    // 2D Array for head-to-head comparisons between teams
    private int[][] headToHead;

    public DataStore(int numTeams) {
        this.gameResults = new ArrayList<>();
        this.teamStatistics = new ArrayList<>();
        this.teamDataMap = new HashMap<>();
        this.gameResultQueue = new LinkedList<>();
        this.teamRanking = new PriorityQueue<>(Comparator.comparingDouble(TeamData::getSeedingScore).reversed());
        this.headToHead = new int[numTeams][numTeams];
    }

    public void addGameResult(String result) {
        gameResults.add(result);
        gameResultQueue.offer(result);  // Add to real-time queue for immediate processing
    }

    public void addTeamStatistic(String statistic) {
        teamStatistics.add(statistic);
    }

    public void addTeamRecord(String teamName, TeamRecord record) {
        teamDataMap.computeIfAbsent(teamName, k -> new ArrayList<>()).add(record);
    }

    public void updateHeadToHead(int teamIndex1, int teamIndex2, int result) {
        headToHead[teamIndex1][teamIndex2] = result;
    }

    public void displayTeamData(String teamName) {
        List<TeamRecord> records = teamDataMap.get(teamName);
        if (records != null) {
            System.out.println("Data for team: " + teamName);
            for (TeamRecord record : records) {
                System.out.println(record);
            }
        } else {
            System.out.println("No data found for team: " + teamName);
        }
    }

    public void displayRankedTeams() {
        System.out.println("Team Rankings based on Seeding Score:");
        PriorityQueue<TeamData> tempRanking = new PriorityQueue<>(teamRanking);
        while (!tempRanking.isEmpty()) {
            System.out.println(tempRanking.poll());
        }
    }

    // Define any additional methods here as needed to interact with these structures

    // Inner class to hold each team's yearly record
    public static class TeamRecord {
        private int rank;
        private double statistic;
        private int year;

        public TeamRecord(int rank, double statistic, int year) {
            this.rank = rank;
            this.statistic = statistic;
            this.year = year;
        }

        @Override
        public String toString() {
            return "Year: " + year + ", Rank: " + rank + ", Statistic: " + statistic;
        }
    }

    // Inner class to hold team-specific data, primarily for ranking purposes
    public static class TeamData {
        private String teamName;
        private double seedingScore;  // A score used for ranking purposes
        private int pointsScored;
        private int wins;
        private int losses;

        public TeamData(String teamName, double seedingScore, int pointsScored, int wins, int losses) {
            this.teamName = teamName;
            this.seedingScore = seedingScore;
            this.pointsScored = pointsScored;
            this.wins = wins;
            this.losses = losses;
        }

        public double getSeedingScore() {
            return seedingScore;
        }

        @Override
        public String toString() {
            return "Team: " + teamName + ", Seeding Score: " + seedingScore +
                   ", Points: " + pointsScored + ", Wins: " + wins + ", Losses: " + losses;
        }
    }
    public static void main(String[] args) {
        // Initialize DataStore (assume 30 teams)
        DataStore dataStore = new DataStore(30);

        try {
            String dataFolderPath = "/content/Cleaned_Data";  // Path to main data folder
            File dataFolder = new File(dataFolderPath);

            // Check if directory exists
            if (!dataFolder.exists() || !dataFolder.isDirectory()) {
                System.out.println("Directory not found: " + dataFolderPath);
                return;
            }

            // Iterate through each category folder in the main directory
            for (File categoryFolder : dataFolder.listFiles()) {
                if (categoryFolder.isDirectory()) {
                    System.out.println("Category: " + categoryFolder.getName());

                    // Iterate through each team CSV file in the category folder
                    for (File teamFile : categoryFolder.listFiles()) {
                        if (teamFile.isFile() && teamFile.getName().endsWith(".csv")) {
                            // Load data for each team
                            DataSource source = new DataSource(teamFile.getPath());
                            Instances data = source.getDataSet();

                            // Use file name (without extension) as team name
                            String teamName = teamFile.getName().replace(".csv", "");

                            // Process each row (instance) in the data
                            for (int i = 0; i < data.numInstances(); i++) {
                                Instance instance = data.instance(i);

                                // Assuming the attributes are in the order: Rank, Statistic, Year
                                int rank = (int) instance.value(0); // Rank
                                double statistic = instance.value(1); // Statistic
                                int year = (int) instance.value(2); // Year

                                // Create a new TeamRecord for each year
                                DataStore.TeamRecord record = new DataStore.TeamRecord(rank, statistic, year);

                                // Add the record to the DataStore
                                dataStore.addTeamRecord(teamName, record);
                            }

                            System.out.println("Loaded and stored data for team: " + teamName);
                        }
                    }
                }
            }

            // Display all stored data for a sample team, e.g., "Phoenix"
            dataStore.displayTeamData("Phoenix");

        } catch (Exception e) {
            e.printStackTrace();
        }
    }
}
```

### Features for NBA Game Prediction Model

- **Average Scoring Margin** (`average_scoring_margin`): Difference between points scored and points allowed, on average.
- **Defensive Efficiency** (`defensive_efficiency`): Points allowed per 100 possessions.
- **Effective Field Goal Percentage** (`efg_pct`): Adjusts field goal percentage to account for three-point shots.
- **Floor Percentage** (`flr_pct`): Percentage of possessions resulting in at least one point.
- **Opponent Floor Percentage** (`opp_flr_pct`): Opponent's percentage of possessions resulting in at least one point.
- **Opponent Effective Field Goal Percentage** (`opp_efg_pct`): Adjusted field goal percentage for opponent, considering three-point shots.
- **Shooting Percentage** (`sht_pct`): Percentage of successful field goals made.
- **Turnover Percentage** (`tov_pct`): Percentage of possessions ending in a turnover.
